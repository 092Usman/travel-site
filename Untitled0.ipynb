```python
import time
import json
!pip install selenium
from selenium.webdriver.support.wait import WebDriverWait
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
 


class TestNews():
    def setup_method(self, method):
        self.driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
        self.vars = {}

    def teardown_method(self, method):
        self.driver.quit()

    def test_news(self):
        # Test name: news
        # Step # | name | target | value | comment
        # 1 | open | /news |  |
        self.driver.get("https://www.tapmad.com/news")
        # 2 | setWindowSize | 720x720 |  |
        self.driver.set_window_size(720, 720)
        # 3 | waitForElementPresent | xpath=//img[@alt='News'] | 3000 |
        WebDriverWait(self.driver, 3).until(
            expected_conditions.presence_of_element_located((By.XPATH, "//img[@alt=\'News\']")))
        # 4 | mouseOver | css=.usr_btns > .nav-item:nth-child(3) > a |  |
        element = self.driver.find_element(
            By.CSS_SELECTOR, ".usr_btns > .nav-item:nth-child(3) > a")
        actions = ActionChains(self.driver)
        actions.move_to_element(element).perform()
        # 5 | storeAttribute | xpath=//div[@id='__next']/div/div[2]/div[2]/div[2]/div[2]/a/div/div/img@src | imgurl |
        attribute = self.driver.find_element(
            By.XPATH, "//div[@id=\'__next\']/div/div[2]/div[2]/div[2]/div[2]/a/div/div/img").get_attribute("src")
        self.vars["imgurl"] = attribute
        # 6 | echo | ${imgurl} |  |
        print("{}".format(self.vars["imgurl"]))
        # 7 | click | xpath=//img[@alt='News'] |  |
        self.driver.find_element(By.XPATH, "//img[@alt=\'News\']").click()
        # 8 | waitForElementPresent | xpath=//div[@id='__next']/div/div[2]/div[2]/div[2]/div/div/div/h4 | 10000 |
        WebDriverWait(self.driver, 10).until(expected_conditions.presence_of_element_located(
            (By.XPATH, "//div[@id=\'__next\']/div/div[2]/div[2]/div[2]/div/div/div/h4")))
        # 9 | storeText | xpath=//div[@id='__next']/div/div[2]/div[2]/div[2]/div/div/div/h4 | heading |
        self.vars["heading"] = self.driver.find_element(
            By.XPATH, "//div[@id=\'__next\']/div/div[2]/div[2]/div[2]/div/div/div/h4").text
        # 10 | echo | ${heading} |  |
        print("{}".format(self.vars["heading"]))
        # 11 | waitForElementPresent | xpath=//div[@id='__next']/div/div[2]/div[2]/div[2]/div/div/div/p | 3000 |
        WebDriverWait(self.driver, 3).until(expected_conditions.presence_of_element_located(
            (By.XPATH, "//div[@id=\'__next\']/div/div[2]/div[2]/div[2]/div/div/div/p")))
        # 12 | storeText | xpath=//div[@id='__next']/div/div[2]/div[2]/div[2]/div/div/div/p | newstext |
        self.vars["newstext"] = self.driver.find_element(
            By.XPATH, "//div[@id=\'__next\']/div/div[2]/div[2]/div[2]/div/div/div/p").text
        # 13 | echo | ${newstext} |  |
        print("{}".format(self.vars["newstext"]))
        # 14 | echo | ${imgurl} |  |
        print("{}".format(self.vars["imgurl"]))

# compare with last news
        # Opening JSON file
        f = open("/content/drive/MyDrive/Colab Notebooks/newsList.json", encoding="utf-8")

        # returns JSON object as a dictionary
        data = json.load(f)

        heading = data["Heading"]

        if(heading == self.vars["heading"]):
            print("Already Posted news")

            # Last | close |  |  |
            self.driver.close()

            # Closing file
            f.close()

        else:

          # swith to streamify for posting

            # Test name: postNews
            # Step # | name | target | value | comment
            # 1 | open | /wp-admin |  |
            self.driver.get("https://streamify.cf/wp-admin")
            # 2 | setWindowSize | 720x720 |  |
            self.driver.set_window_size(720, 720)
            # 3 | storeTitle | / | title |
            self.vars["title"] = self.driver.title
            # 4 | if | ${title}==="Log In ‹ Streamify — WordPress" |  |
            if self.driver.execute_script("return (arguments[0]===\"Log In ‹ Streamify — WordPress\")", self.vars["title"]):
                # 5 | type | id=user_login | admin |
                self.driver.find_element(
                    By.ID, "user_login").send_keys("admin")
                # 6 | type | id=user_pass | usman786 |
                self.driver.find_element(
                    By.ID, "user_pass").send_keys("usman786")
                # 7 | click | id=wp-submit |  |
                self.driver.find_element(By.ID, "wp-submit").click()
                # 8 | end |  |  |
            # 9 | click | css=#wp-admin-bar-menu-toggle .ab-icon |  |
            self.driver.find_element(
                By.CSS_SELECTOR, "#wp-admin-bar-menu-toggle .ab-icon").click()
            # 10 | click | css=.menu-icon-post > .wp-menu-name |  |
            self.driver.find_element(
                By.CSS_SELECTOR, ".menu-icon-post > .wp-menu-name").click()
            # 11 | click | linkText=Add New |  |
            self.driver.find_element(By.LINK_TEXT, "Add New").click()

    # Skip guide
            # 1 | waitForElementPresent | css=.components-guide__forward-button | 30000 |
            WebDriverWait(self.driver, 30).until(expected_conditions.presence_of_element_located(
                (By.CSS_SELECTOR, ".components-guide__forward-button")))

            # 2 | click | for skiping starting guide
            for i in range(3):
                self.driver.find_element(
                    By.CSS_SELECTOR, ".components-guide__forward-button").click()

            self.driver.find_element(
                By.CSS_SELECTOR, ".components-guide__finish-button").click()
    # end

            # 19 | click | css=.components-dropdown-menu__toggle > svg |  |
            self.driver.find_element(
                By.CSS_SELECTOR, ".components-dropdown-menu__toggle > svg").click()
            # 20 | click | css=.components-menu-items-choice:nth-child(2) |  |
            self.driver.find_element(
                By.CSS_SELECTOR, ".components-menu-items-choice:nth-child(2)").click()
            # 21 | click | css=.components-dropdown-menu__toggle > svg |  |
            self.driver.find_element(
                By.CSS_SELECTOR, ".components-dropdown-menu__toggle > svg").click()
            # 22 | click | xpath=//textarea |  |
            self.driver.find_element(By.XPATH, "//textarea").click()
            # 23 | type | xpath=//textarea | ${heading} |
            self.driver.find_element(
                By.XPATH, "//textarea").send_keys(self.vars["heading"])
            # 24 | click | xpath=//div[2]/textarea |  |
            self.driver.find_element(By.XPATH, "//div[2]/textarea").click()
            # 25 | type | xpath=//div[2]/textarea | ${newstext} |
            self.driver.find_element(
                By.XPATH, "//div[2]/textarea").send_keys(self.vars["newstext"])
            # 26 | click | css=.interface-pinned-items path |  |
            self.driver.find_element(
                By.CSS_SELECTOR, ".interface-pinned-items > .components-button").click()
            # 27 | click | css=.toggle-indicator |  |
        #    self.driver.find_element(By.CSS_SELECTOR, ".toggle-indicator").click()
            # 28 | click | xpath=//table[3]/tbody/tr/td/input |  |
            self.driver.find_element(
                By.XPATH, "//*[@id=\"fifu_input_url\"]").click()
            # 29 | type | xpath=//table[3]/tbody/tr/td/input | ${imgurl} |
            self.driver.find_element(
                By.XPATH, "//*[@id=\"fifu_input_url\"]").send_keys(self.vars["imgurl"])
            # 30 | click | xpath=//form/div/div/div/div/div[2]/div/a |  |
            self.driver.find_element(
                By.XPATH, "//form/div/div/div/div/div[2]/div/a").click()
            # 31 | waitForElementPresent | xpath=//form/div/div/div/div/div[2]/div/div[3] | 30000 |
            WebDriverWait(self.driver, 30).until(expected_conditions.presence_of_element_located(
                (By.XPATH, "//form/div/div/div/div/div[2]/div/div[3]")))
            # 32 | click | css=.interface-complementary-area-header__small path |  |
            self.driver.find_element(
                By.CSS_SELECTOR, ".interface-complementary-area-header__small path").click()
            # 33 | click | css=.editor-post-publish-panel__toggle |  |
            self.driver.find_element(
                By.CSS_SELECTOR, ".editor-post-publish-panel__toggle").click()
            # 34 | click | css=.editor-post-publish-button |  |
            self.driver.find_element(
                By.CSS_SELECTOR, ".editor-post-publish-button").click()
            # 35 | waitForElementPresent | xpath=//div[@id='editor']/div/div/div[2]/div[4]/div[2]/div/div/div[2]/div/div[2]/div/div/input | 30000 |
            WebDriverWait(self.driver, 30).until(expected_conditions.presence_of_element_located(
                (By.XPATH, "//div[@id=\'editor\']/div/div/div[2]/div[4]/div[2]/div/div/div[2]/div/div[2]/div/div/input")))
            # 36 | storeText | xpath=//div[@id='editor']/div/div/div[2]/div[4]/div[2]/div/div/div[2]/div/div[2]/div/div/input | postlink |
            self.vars["postlink"] = self.driver.find_element(
                By.XPATH, "//div[@id=\'editor\']/div/div/div[2]/div[4]/div[2]/div/div/div[2]/div/div[2]/div/div/input").text
            # 37 | echo | ${postlink} |  |
            print("{}".format(self.vars["postlink"]))

            # Last | close |  |  |
            self.driver.close()

            # Closing file OPNED FOR READING
            f.close()

    # Writing news to json
            News = {
                "Heading": self.vars["heading"],
                "Image Url": self.vars["imgurl"],
                "news": self.vars["newstext"],
            }

            # Serializing json
            json_object = json.dumps(News, indent=3, ensure_ascii=False)

            # Writing to sample.json
            with open("/content/sample_data/newsList.json", "w", encoding="utf-8") as outfile:
                outfile.write(json_object)


# For printing source code of page
# print(driver.page_source)

while True:
    testClass = TestNews()
    testClass.setup_method("")
    testClass.test_news()
    testClass.teardown_method("")
    time.sleep(60*15)
```
